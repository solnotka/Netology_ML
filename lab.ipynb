```python
from sklearn.preprocessing import OneHotEncoder

import pandas as pd
import re
```

```python
cols = []
with open ('flag.names', 'r') as f:
        for line in f.readlines():
            l = list(line.split())
            if l and re.match(r'\d+\.', l[0]) and l[1][0].islower():
                cols.append(l[1])

cols
```

```python
data = pd.read_csv('flag.data', header=None, names=cols)
data
```

```python
data_flags = data.iloc[:, 6:]
enc = OneHotEncoder()
enc.fit(data_flags[['mainhue', 'topleft', 'botright']])
df = enc.transform(data_flags[['mainhue', 'topleft', 'botright']]).toarray()
df.shape
```

```python
dfd = pd.DataFrame(data=df, columns=df_columns)
dfd
```
